```python
%matplotlib inline
import pandas as pd
import pydataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pydataset.data('titanic')
```

```python
titanic.head()
```

```python
titanic['class'].value_counts()
```

```python
titanic['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
titanic['survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
titanic.groupby('survived')['class'].value_counts()
```

```python
titanic.groupby('survived')['class'].value_counts().plot(kind='bar')
```

```python
estados = pd.read_csv('./datasets/populacao_brasil.csv')
```

```python
estados.head()
```

```python
estados['total'].hist()#Histograma
plt.show() #Representa gráfico em notação científica
```

```python
fig, ax = plt.subplots()
estados['total'].hist(bins=15, orientation='horizontal') #Bins é o numero de colunas
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['percent'] = estados['total']/estados['total'].sum()
```

```python
estados.head()
```

```python
plt.pie(estados['percent'], labels=estados['estado'])#Gráfico de pizza
plt.show()
```

```python
plt.pie(estados['percent'], labels=estados['estado'], autopct='%1.2f%%')#Gráfico de pizza
plt.show()
```
