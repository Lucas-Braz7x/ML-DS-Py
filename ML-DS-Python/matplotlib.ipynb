```python
%matplotlib inline
import matplotlib.pyplot as plt #Importando e chamando de 'plt'
import numpy as np
#import pandas as pd
```

```python
x = np.linspace(1,10,100)#Emite um tipo de array
x
```

```python
plt.plot(x, np.sin(x), 'g--')
```
