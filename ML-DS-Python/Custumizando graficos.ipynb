```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./datasets/reported.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.fillna(0, inplace=True)#Substitui 'NaN'
df.head()
```

```python
plt.plot(df['Year'], df['crimes.total'], '-r')
plt.plot(df['Year'], df['crimes.person'], '-b')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r',label='Total de crimes')
ax.plot(df['Year'], df['crimes.person'], '-b',label='Crimes.Person')
ax.legend()
ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')
ax.set_title('Crimes: Total x Person')
ax.set_xlim([df['Year'].min(), df['Year'].max()])
ax.plot()
```
